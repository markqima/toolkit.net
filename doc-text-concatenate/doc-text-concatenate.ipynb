```python
from docx import Document
import pandas as pd


def get_text(filename):
    doc = Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text.strip())
    return ';'.join(fullText)
```

```python
import os

folder = 'D:/data/26109137-2020-12-11/'
result_df = pd.DataFrame()
total = 0
for file_ in os.listdir(folder):
    if file_.endswith('docx'):
        line_text = get_text(folder+file_)
        if ';' in line_text:
            current = str.split(line_text, ';')
        else:
            current = str.split(line_text, '\n')
        current_len = len(current)
        total+=current_len
        df = pd.DataFrame(current)
#        print(file_,line_text, current, current_len)
    if file_.endswith('xlsx'):
        df = pd.read_excel(folder+file_, header = None)
        df[0] = df[0].str[:-1]
        # print(file_,line_text, df)
    result_df = pd.concat([result_df, df])

result_df
```

```python
result_df.replace("", float("NaN"), inplace=True)
result_df.dropna(inplace=True)
result_df.to_csv('./result.csv', encoding='utf_8_sig', index=False)
result_df
```
